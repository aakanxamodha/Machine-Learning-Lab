```python
'''
6) Prepare and excel sheet for Student's data with the following content: 
(a) data rows of minimum 10 students data. 
(b) Columns Data: SNO, REGNO, NAME, M1, M2, M3, M4, M5. 
(c) Computer the Mean, Median, Mode, Average and 5 number summary of all Marks Columns 
(d) Draw a box plot for all numeric columns. 
(e) Draw various possible plots for student marks in each subject. 
(f) Display the outlier values if any in each of the students marks columns.

'''
import pandas as pd
import names
import random

def generate_student_data(num_students):
    data = []
    for i in range(1, num_students + 1):
        sno = i
        regno = f"REG{1000 + i}"
        name = names.get_full_name()
        marks = [random.randint(0, 100) for _ in range(5)]  # Generating marks for 5 subjects
        student_data = [sno, regno, name] + marks
        data.append(student_data)
    return data
    
num_students = 10
columns = ['SNO', 'REGNO', 'NAME', 'M1', 'M2', 'M3', 'M4', 'M5']
student_data = generate_student_data(num_students)
df = pd.DataFrame(student_data, columns=columns)
df.to_excel('students_data.xlsx', index=False, engine='openpyxl')
print("Excel sheet 'students_data.xlsx' has been created.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_excel('students_data.xlsx', engine='openpyxl')

marks_columns = ['M1', 'M2', 'M3', 'M4', 'M5']

mean_dict = {}
median_dict = {}
mode_dict = {}
five_number_summary_dict = {}

for column in marks_columns:
    data = df[column].dropna() 
    
    mean = np.mean(data)
    median = np.median(data)
    
    mode_result = stats.mode(data, keepdims=True)
    mode = mode_result.mode[0]  
    
    five_number_summary = np.percentile(data, [0, 25, 50, 75, 100])
    
    mean_dict[column] = mean
    median_dict[column] = median
    mode_dict[column] = mode
    five_number_summary_dict[column] = five_number_summary

print("Mean:")
for column in mean_dict:
    print(column, ":", mean_dict[column])

print("Median:")
for column in median_dict:
    print(column, ":", median_dict[column])

print("Mode:")
for column in mode_dict:
    print(column, ":", mode_dict[column])

print("5-Number Summary:")
for column in five_number_summary_dict:
    print(column, ":", five_number_summary_dict[column])

plt.figure(figsize=(12, 8))
plt.boxplot([df[col].dropna() for col in marks_columns], tick_labels=marks_columns, patch_artist=True)
plt.title('Box Plot of Marks')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()

for column in marks_columns:
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    plt.hist(df[column].dropna(), bins=10, edgecolor='black', color='lightblue')
    plt.title(column + ' Histogram')
    plt.xlabel('Marks')
    plt.ylabel('Frequency')
    
    plt.subplot(1, 3, 2)
    plt.boxplot(df[column].dropna(), vert=False, patch_artist=True)
    plt.title(column + ' Box Plot')
    plt.xlabel('Marks')
    
    plt.subplot(1, 3, 3)
    plt.violinplot(df[column].dropna(), showmeans=True)
    plt.title(column + ' Violin Plot')
    plt.xlabel('Marks')
    
    plt.tight_layout()
    plt.show()

def find_outliers(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

print("Outliers in each subject:")
for column in marks_columns:
    outliers = find_outliers(df[column].dropna())
    if not outliers.empty:
        print(column, "Outliers:")
        print(outliers.values)
    else:
        print(column, "No outliers")
```

```python
'''
8) Write a program in Python to create a new csv file and fill in the data used in Q.4 and display the .csv file. 

'''

import numpy as np
import csv

M = int(input("Enter the number of rows (M): "))
N = int(input("Enter the number of columns (N): "))

matrix = np.random.randint(1, 10, (M, N))
print("Matrix:")
print(matrix)

matrix_addition = matrix + matrix
print("Matrix Addition:")
print(matrix_addition)

if M == N:
    matrix_multiplication = np.dot(matrix, matrix)
    print("Matrix Multiplication:")
    print(matrix_multiplication)
else:
    print("Matrix Multiplication: Not possible for non-square matrices")

if M == N:
    try:
        matrix_inversion = np.linalg.inv(matrix)
        print("Matrix Inversion:")
        print(matrix_inversion)
    except np.linalg.LinAlgError:
        matrix_inversion = None
        print("Matrix Inversion: Matrix is singular and cannot be inverted")
else:
    matrix_inversion = None
    print("Matrix Inversion: Not possible for non-square matrices")

with open('matrix_operations.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    
    writer.writerow(["Original Matrix"])
    writer.writerows(matrix)
    
    writer.writerow(["Matrix Addition"])
    writer.writerows(matrix_addition)
    
    if M == N:
        writer.writerow(["Matrix Multiplication"])
        writer.writerows(matrix_multiplication)
        
        if matrix_inversion is not None:
            writer.writerow(["Matrix Inversion"])
            writer.writerows(matrix_inversion)
        else:
            writer.writerow(["Matrix Inversion"])
            writer.writerow(["Matrix is singular and cannot be inverted"])
    else:
        writer.writerow(["Matrix Multiplication"])
        writer.writerow(["Not possible"])
        writer.writerow(["Matrix Inversion"])
        writer.writerow(["Not possible"])

print("CSV file 'matrix_operations.csv' created and data has been written.")
```
